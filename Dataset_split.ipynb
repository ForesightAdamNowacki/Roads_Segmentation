```python
import pandas as pd
import numpy as np
import os
import shutil
```

```python
dataset_path = "D:/GitHub/Datasets/MassachusettsRoads"
images_crops_verified_path = os.path.join(dataset_path, "Images_Crops_Verified")
masks_crops_verified_path = os.path.join(dataset_path, "Masks_Crops_Verified")

train_dir = os.path.join(dataset_path, "Train")
train_images_dir = os.path.join(train_dir, "Images")
train_masks_dir = os.path.join(train_dir, "Masks")

validation_dir = os.path.join(dataset_path, "Validation")
validation_images_dir = os.path.join(validation_dir, "Images")
validation_masks_dir = os.path.join(validation_dir, "Masks")

if not os.path.exists(train_dir):
	os.mkdir(train_dir)
	os.mkdir(train_images_dir)
	os.mkdir(train_masks_dir)
    
if not os.path.exists(validation_dir):
	os.mkdir(validation_dir)
	os.mkdir(validation_images_dir)
	os.mkdir(validation_masks_dir)

def listdir_fullpath(path):
    return [os.path.join(path, file) for file in os.listdir(path)]

train_size = 0.8
validation_size = 1 - train_size
```

```python
# Images:
images = listdir_fullpath(path = images_crops_verified_path)
id = int(len(images) * train_size)
train_images_from = images[:id]
validation_images_from = images[id:]

# Train images:
for i in np.arange(0, len(train_images_from)):
    from_ = train_images_from[i]
    to_ = from_.replace("Images_Crops_Verified", "Train\\Images")
    shutil.copyfile(from_, to_)
    print(i, "From:", from_, "To:", to_)
    
# Validation images:
for i in np.arange(0, len(validation_images_from)):
    from_ = validation_images_from[i]
    to_ = from_.replace("Images_Crops_Verified", "Validation\\Images")
    shutil.copyfile(from_, to_)
    print(i, "From:", from_, "To:", to_)
```

```python
# Masks:
masks = listdir_fullpath(path = masks_crops_verified_path)
id = int(len(masks) * train_size)
train_masks_from = masks[:id]
validation_masks_from = masks[id:]

# Train masks:
for i in np.arange(0, len(train_masks_from)):
    from_ = train_masks_from[i]
    to_ = from_.replace("Masks_Crops_Verified", "Train\\Masks")
    shutil.copyfile(from_, to_)
    print(i, "From:", from_, "To:", to_)
    
# Validation masks:
for i in np.arange(0, len(validation_masks_from)):
    from_ = validation_masks_from[i]
    to_ = from_.replace("Masks_Crops_Verified", "Validation\\Masks")
    shutil.copyfile(from_, to_)
    print(i, "From:", from_, "To:", to_)
```

```python
if os.listdir(train_images_dir) == os.listdir(train_masks_dir):
    print('Lists are exactly')
else:
    print('Lists are not equal')

if os.listdir(validation_images_dir) == os.listdir(validation_masks_dir):
    print('Lists are exactly')
else:
    print('Lists are not equal')
```
