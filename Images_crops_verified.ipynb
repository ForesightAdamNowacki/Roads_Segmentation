```python
import numpy as np
import os
import cv2
```

```python
dataset_path = "D:/GitHub/Datasets/MassachusettsRoads"
images_crops_path = os.path.join(dataset_path, "Images_Crops")
masks_crops_path = os.path.join(dataset_path, "Masks_Crops")
images_crops_verified_path = os.path.join(dataset_path, "Images_Crops_Verified")
masks_crops_verified_path = os.path.join(dataset_path, "Masks_Crops_Verified")

if not os.path.exists(images_crops_verified_path):
	os.mkdir(images_crops_verified_path)
if not os.path.exists(masks_crops_verified_path):
	os.mkdir(masks_crops_verified_path)

def listdir_fullpath(path):
    return [os.path.join(path, file) for file in os.listdir(path)]

images = listdir_fullpath(path = images_crops_path)
len(images)
```

```python
threshold = 0.01

for i in np.arange(0, len(images)):
    image = cv2.imread(images[i])
    
    image_ = np.concatenate([np.expand_dims(image[:,:,0] >= 255, -1),
                             np.expand_dims(image[:,:,1] >= 255, -1),
                             np.expand_dims(image[:,:,2] >= 255, -1)], axis = -1)
    image_ = np.sum(image_, axis = -1)
    blanks = np.sum(image_ == 3)
    image_dim = image.shape[0] * image.shape[1]
    ratio = blanks/image_dim
    
    if ratio > threshold:
        print(i, "Image will be dropped:", images[i], np.round(ratio, 4))
    else:
        print(i, "Image will be saved:", images[i], np.round(ratio, 4))
        new_image_path = os.path.join(images_crops_verified_path, os.path.basename(images[i]))
        cv2.imwrite(new_image_path, image)
```

```python
images_crops_verified = listdir_fullpath(path = images_crops_verified_path)
masks_crops_verified = [i.replace("Images_Crops_Verified", "Masks_Crops") for i in images_crops_verified]
print("Accepted:", len(masks_crops_verified)/len(images))
```

```python
for i in np.arange(0, len(masks_crops_verified)):
    image = cv2.imread(masks_crops_verified[i])
    mask_verified_path = masks_crops_verified[i].replace("Masks_Crops", "Masks_Crops_Verified")
    cv2.imwrite(mask_verified_path, image)
    print("Save image mask:", mask_verified_path)
```

```python
images_crops_verified_ = os.listdir(images_crops_verified_path)
masks_crops_verified_ = os.listdir(masks_crops_verified_path)

if images_crops_verified_ == masks_crops_verified_:
    print('Lists are exactly')
else:
    print('Lists are not equal')
```
