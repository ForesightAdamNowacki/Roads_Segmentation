```python
import numpy as np
import os
import cv2
```

```python
repository_dir = "D:/GitHub/Roads_segmentation"
dataset_path = "D:/GitHub/Datasets/MassachusettsRoads"
images_path = os.path.join(dataset_path, "Images")
masks_path = os.path.join(dataset_path, "Masks")
images_verified_path = os.path.join(dataset_path, "Images_Verified")
masks_verified_path = os.path.join(dataset_path, "Masks_Verified")

if not os.path.exists(images_verified_path):
	os.mkdir(images_verified_path)
if not os.path.exists(masks_verified_path):
	os.mkdir(masks_verified_path)

def listdir_fullpath(path):
    return [os.path.join(path, file) for file in os.listdir(path)]

images = listdir_fullpath(path = images_path)
```

```python
threshold = 0.5

for i in np.arange(0, len(images)):
    image = cv2.imread(images[i])
    
    image_ = np.concatenate([np.expand_dims(image[:,:,0] >= 255, -1),
                             np.expand_dims(image[:,:,1] >= 255, -1),
                             np.expand_dims(image[:,:,2] >= 255, -1)], axis = -1)
    image_ = np.sum(image_, axis = -1)
    blanks = np.sum(image_ == 3)
    image_dim = image.shape[0] * image.shape[1]
    ratio = blanks/image_dim
    
    if ratio > threshold:
        print(i, "Image will be dropped:", images[i], np.round(ratio, 4))
    else:
        print(i, "Image will be saved:", images[i], np.round(ratio, 4))
        new_image_path = os.path.join(images_verified_path, os.path.basename(images[i]))
        cv2.imwrite(new_image_path, image)
```

```python
images_verified = listdir_fullpath(path = images_verified_path)
images_verified_basename = [i.replace("Images_Verified", "Masks") for i in images_verified]

for i in np.arange(0, len(images_verified_basename)):
    image = cv2.imread(images_verified_basename[i])
    mask_verified_path = images_verified_basename[i].replace("Masks", "Masks_Verified")
    cv2.imwrite(mask_verified_path, image)
    print("Save image mask:", mask_verified_path)
```

```python
images_verified_ = [os.path.basename(i) for i in images_verified_basename]
masks_verified_ = os.listdir(masks_verified_path)

if images_verified_ == masks_verified_:
    print('Lists are exactly')
else:
    print('Lists are not equal')
```
