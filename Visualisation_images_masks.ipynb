```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset_path = "D:/GitHub/Datasets/MassachusettsRoads"
images = os.path.join(dataset_path, "Images_Crops_Verified")
masks = os.path.join(dataset_path, "Masks_Crops_Verified")

id = 13479
print(os.path.join(images, os.listdir(images)[id]))
background = cv2.imread(os.path.join(images, os.listdir(images)[id]))
overlay = cv2.imread(os.path.join(masks, os.listdir(masks)[id]))
added_image = cv2.addWeighted(background, 1, overlay, 1, 1)

fig = plt.figure(figsize = (20, 10))
ax1 = fig.add_subplot(1, 3, 1)
ax1.imshow(background)
ax2 = fig.add_subplot(1, 3, 2)
ax2.imshow(overlay)
ax3 = fig.add_subplot(1, 3, 3)
ax3.imshow(added_image)
```
